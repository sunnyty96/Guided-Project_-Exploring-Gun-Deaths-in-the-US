```python
import csv
f=open("guns.csv")
csvreader=csv.reader(f)
data=list(csvreader)
print(data[:6])
```

```python
headers=data[0]
data=data[1:]
print(headers)
print(data[:5])
```

```python
years=[]
for row in data:
    years.append(row[1])
print(years)
```

```python
year_counts={}
for item in years:
    if item in year_counts:
        year_counts[item]+=1
    else:
        year_counts[item]=1
print(year_counts)
```
